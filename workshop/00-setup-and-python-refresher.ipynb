## Step Three: Opening files

```python
import os
from PIL import Image, ImageOps
import numpy as np
```

```python
data_path = '/content/gdrive/MyDrive/lucerne-digital-skills/data/'
```

```python
protest_path = '{}protest/test/negative/'.format(data_path)
```

```python
os.listdir(protest_path)
```

```python
for item in os.listdir(protest_path):
  print("Filename:", item) #Show the name of the file
  img_path = os.path.join(protest_path, item)
  img = Image.open(img_path) #Open the image
  print("Size:", img.size, '\n') #Print the size of the image (in pixels)
```
